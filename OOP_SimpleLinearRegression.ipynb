```python
class Simple_Linear_Regression :
    # Attributes  if any 
    def __init__(self,X_train,X_test,y_train,y_test):
        #  self.X_train= X_train    ... do it for all 
        self.X_train= X_train
        self.X_test= X_test
        self.y_train= y_train
        self.y_test= y_test
        
    def Mean(self,m):
        return( np.sum(m)/len(m))
    
    def Variance(self,m):
        return(np.sum((m-self.Mean(m))**2)/len(m))
        
    def Covariance(self,m):
        return (self.Variance(m)**0.5)
    
    def Coefficents(self,x,y):
        w1=(np.sum((x-self.Mean(x))*(y-self.Mean(y)))/np.sum((x-self.Mean(x))**2))
        w0=self.Mean(y)-w1*self.Mean(x)
        return w0,w1
    
    def simple_linear_regression(self):
        w0,w1=self.Coefficents(self.X_train,self.y_train)
        preds=w0 + w1*self.X_test
        return( preds)
    
    def MSE(self,actual, predicted):
        return(np.sum((actual-predicted)**2)/len(actual))

    def Evaluate():
         return(self.MSE(self.simple_linear_regression(),self.y_test))
    def Vizualize(self):
        plt.figure(figsize=(10,8))
        plt.title('PREDICTION VS ACTUAL ', fontsize=24 , fontstyle='italic')
        plt.plot( self.X_test, self.simple_linear_regression(), linewidth=2.0)  ## REPLACE X AND Y WITH  X_test AND PREDS ARRAYS! 
#plt.scatter(X_train ,y_train , alpha =0.5 ,label='Y')  # NOTE : Make Sure  x_train , y_train  IN Data Type ( Array )
        plt.scatter(self.X_test , self.y_test , alpha =0.5 ,label='Y')  # NOTE : Make Sure  x_train , y_train  IN Data Type ( Array )
        plt.show
                    # You can make this on your own by plotting  Points and Line 
                        # plt.scatter () , plt.plot()
         
```
